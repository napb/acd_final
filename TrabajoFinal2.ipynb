```python
def connectAccount(user, client):
    return client.search_recent_tweets(
        query='from:'+user,
        tweet_fields=[],
        max_results=100
    )
```

```python
stop_words = list(stopwords.words('spanish'))
stop_words += [',', '.', ':', ';', 'üì¢', 'rt', 'üìö', '‚Ä¶', '..', '‚ú®', '!', '¬°', '-', 'üìª', 'ü§©', 'üîä', 'üòÅ', 'ver', '‚úçÔ∏è', '‚Äô', '}', '(', ')', '‚ö°Ô∏è', '‚åõÔ∏è‚ú®', 'ü•á', 'üì≥', 'üÜï', 'üëá', '"', "'", '‚Äù', '?', '¬ø', '']
```

```python
def analizeTweets(tweets) :
    d ={}
    try:
        for tweet in tweets.data:
            tk = ToktokTokenizer().tokenize(tweet.text.lower())
            filtered_words =[w for w in tk if not w in stop_words]

            for w in filtered_words:
                if w in d:
                    d[w] += 1
                else:
                    d[w] = 1
    except:
        pass

    return d
```

```python
def sortDict(d):
    sorted_counter = sorted(d, key=d.get, reverse=True)
    for w in sorted_counter[:100]:
        print("conteo:", d[w], "\t|\tpalabra:'", w, "'")
```

```python
sortDict(analizeTweets(tweets))
```
